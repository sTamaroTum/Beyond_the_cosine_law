# Validation of the analytical Present model with experimental data (third dataset)

 This is the code to reproduce Figure 20 found in Tamaro et al (in review). 

## Code

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
x = loadmat('exp_data/ini_7_param_Reg3_ExpNumComparison_power.mat')
plt.figure(1,figsize=(9,4))
plt.subplot(2,2,2)
xx = x['x_mod'][1][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][1][0]),np.squeeze(x['err_mod'][1][0]),color='k',fmt='o',barsabove=True,linestyle='',capsize=5,elinewidth=1,label='Exp.',clip_on=False,zorder=1000)
xx = x['x_mod'][0][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][0][0]),np.squeeze(x['err_mod'][0][0]),color='C0',barsabove=True,linestyle='-',capsize=5,elinewidth=1,label='Present model',clip_on=False)
plt.grid()
plt.ylim([0.9,1.1])
plt.xticks(np.linspace(-30,0,4),color='white')
plt.xlim([-30,0])
plt.yticks([0.9,1,1.1],color='w')
plt.title('Mod-TI',fontsize=18)
plt.legend(handlelength=1,loc='upper left',frameon=False,ncol=3,bbox_to_anchor=(-0.45, 1.62),fancybox=True, shadow=False)
plt.subplot(2,2,1)
xx = x['x_mod'][3][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][3][0]),np.squeeze(x['err_mod'][3][0]),color='k',fmt='o',barsabove=True,linestyle='',capsize=5,elinewidth=1,label='Exp.',clip_on=False,zorder=1000)# plt.ylabel('$P/P_{\gamma=0^\circ}$ [-]')
xx = x['x_mod'][2][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][2][0]),np.squeeze(x['err_mod'][2][0]),color='C0',barsabove=True,linestyle='-',capsize=5,elinewidth=1,label='Present model',clip_on=False)
plt.ylabel(r'$\eta_P$ [-]')
plt.grid()
plt.ylim([0.9,1.1])
plt.xticks(np.linspace(-30,30,5),color='white')
plt.xlim([-31,31])
plt.yticks([0.9,1,1.1],color='k')
plt.title('Low-TI',fontsize=18)
plt.figure(1)
plt.tight_layout()
x = loadmat('exp_data/ini_7_param_Reg3_ExpNumComparison_CT.mat')
plt.subplot(2,2,4)
xx = x['x_mod'][1][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][1][0]),np.squeeze(x['err_mod'][1][0]),color='k',fmt='o',barsabove=True,linestyle='',capsize=5,elinewidth=1,label='Exp.',clip_on=False,zorder=1000)
xx = x['x_mod'][0][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][0][0]),np.squeeze(x['err_mod'][0][0]),color='C0',barsabove=True,linestyle='-',capsize=5,elinewidth=1,label='Present model',clip_on=False)
plt.xlabel('$\gamma$ [$^\circ$]')
plt.grid()
plt.ylim([0.9,1.3])
plt.xticks(np.linspace(-30,0,4))
plt.xlim([-30,0])
plt.yticks([0.9,1,1.1,1.2,1.3],color='w')
plt.subplot(2,2,3)
xx = x['x_mod'][3][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][3][0]),np.squeeze(x['err_mod'][3][0]),color='k',fmt='o',barsabove=True,linestyle='',capsize=5,elinewidth=1,label='Exp.',clip_on=False,zorder=1000)# plt.ylabel('$P/P_{\gamma=0^\circ}$ [-]')
xx = x['x_mod'][2][0]
xx= xx.astype(float)
plt.errorbar(np.squeeze(xx),np.squeeze(x['y_mod'][2][0]),np.squeeze(x['err_mod'][2][0]),color='C0',barsabove=True,linestyle='-',capsize=5,elinewidth=1,label='Present model',clip_on=False)
plt.xlabel('$\gamma$ [$^\circ$]')
plt.ylabel(r'$\eta_T$ [-]')
plt.yticks([0.9,1,1.1,1.2,1.3],color='k')
plt.grid()
plt.ylim([0.9,1.3])
plt.xticks(np.linspace(-30,30,5))
plt.xlim([-31,31])
plt.subplot(2,2,1)
plt.text(19.0,1.065,'(a)',fontsize=18)
plt.subplot(2,2,2)
plt.text(-3,1.065,'(b)',fontsize=18)
plt.subplot(2,2,3)
plt.text(19,1.23,'(c)',fontsize=18)
plt.subplot(2,2,4)
plt.text(-3,1.23,'(d)',fontsize=18)
plt.figure(1)
plt.subplots_adjust(left=0.1, bottom=0.175, right=0.95, top=0.85, wspace=0.1, hspace=0.2)
```
