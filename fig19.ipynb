```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
x = loadmat('exp_data/ini_7_param_Reg3_ExpNumComparison_pitch.mat')
```

```python
plt.figure(1,figsize=(9,3.0))
plt.subplot(1,2,2)
xx = x['x_num'][0][0]
xx= xx.astype(float)
plt.plot(np.squeeze(xx),np.squeeze(x['y_num'][0][0]),color='C4',linestyle='',marker='s',label='Mod TI',clip_on=False)
plt.grid()
plt.xticks(np.linspace(-30,0,4),color='k')
plt.xlim([-30,0])
plt.yticks([0,1,2,3,4],color='k')
plt.subplot(1,2,2)
xx = x['x_num'][1][0]
xx= xx.astype(float)
plt.plot(np.squeeze(xx),np.squeeze(x['y_num'][1][0]),color='C3',linestyle='',marker='o',label='Low TI',clip_on=False,zorder=1000)
plt.ylabel(r'$\theta_p$ [$^\circ$]')
plt.grid()
plt.xticks(np.linspace(-30,30,5),color='k')
plt.xlim([-30,30])
plt.yticks([0,2,4,6,8],color='k')
plt.xlabel('$\gamma$ [$^\circ$]')
plt.figure(1)
plt.tight_layout()
```
